```python
#RCNN
import numpy as np
import matplotlib.pyplot as plt
import cv2
import tensorflow as tf
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense, Flatten, Input, TimeDistributed, Dropout
from tensorflow.keras.utils import Sequence
from sklearn.preprocessing import LabelBinarizer
def create_rcnn_model(input_shape):
    base_model = VGG16(weights='imagenet', include_top=False, input_shape=input_shape)
    
    # Define the R-CNN model
    x = Flatten()(base_model.output)
    x = Dense(1024, activation='relu')(x)
    x = Dropout(0.5)(x)
    x = Dense(256, activation='relu')(x)
    output = Dense(num_classes, activation='softmax')(x)  # num_classes is the number of classes in your dataset
    
    model = Model(inputs=base_model.input, outputs=output)
    
    return model
# Function to load images and labels
def load_data(images_path, labels):
    images = []
    for img_path in images_path:
        img = cv2.imread(img_path)
        img = cv2.resize(img, (224, 224))  # Resize to match model input
        images.append(img)
    
    return np.array(images), np.array(labels)

# Example usage
images_path = ["path_to_image1.jpg", "path_to_image2.jpg"]
labels = [0, 1]  # Replace with your labels
X, y = load_data(images_path, labels)
# Define parameters
input_shape = (224, 224, 3)
num_classes = 2  # Update this with the number of classes in your dataset

# Create and compile the model
model = create_rcnn_model(input_shape)
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Convert labels to one-hot encoding
lb = LabelBinarizer()
y = lb.fit_transform(y)

# Train the model
model.fit(X, y, batch_size=16, epochs=10, validation_split=0.2)
def detect_objects(model, image):
    # Preprocess the image for the model
    image_resized = cv2.resize(image, (224, 224))
    image_array = np.expand_dims(image_resized, axis=0)
    
    # Predict
    predictions = model.predict(image_array)
    return predictions

# Example usage
test_image = cv2.imread("path_to_test_image.jpg")
predictions = detect_objects(model, test_image)

# Visualize the results
plt.figure(figsize=(12, 6))
plt.plot(fpr, tpr, color='orange', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve for Object Detection')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.show()
```

```python
import cv2
from ultralytics import YOLO

# Load the YOLOv8 model
model = YOLO('yolov8n.pt')  # You can change this to 'yolov8s.pt', 'yolov8m.pt', etc.

# Function to perform object detection on an image
def detect_objects_in_image(image_path):
    # Read the image
    img = cv2.imread(image_path)

    # Perform inference
    results = model(img)

    # Display results
    results.show()  # Displays the image with bounding boxes and labels

    # Get predictions
    boxes = results.pred[0]  # Predictions for the first image
    return boxes

# Function to perform object detection on video stream
def detect_objects_in_video(video_source=0):
    # Open video source (0 for webcam or path to video file)
    cap = cv2.VideoCapture(video_source)

    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Perform inference
        results = model(frame)

        # Display results
        results.show()  # Displays the frame with bounding boxes and labels

        # Exit on 'q' key press
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

    # Release the capture and destroy windows
    cap.release()
    cv2.destroyAllWindows()

# Main execution
if __name__ == "__main__":
    # Example for image detection
    image_path = 'path/to/your/image.jpg'  # Replace with your image path
    print("Detecting objects in image...")
    detections = detect_objects_in_image(image_path)

    # Print detected bounding boxes and scores
    for box in detections:
        print(f"Class: {box[-1]}, Confidence: {box[-2]:.2f}, Bounding Box: {box[:4]}")

    # Example for video detection (use 0 for webcam)
    print("Detecting objects in video...")
    detect_objects_in_video(video_source=0)  # Use 'path/to/your/video.mp4' for a video file
    plt.figure(figsize=(12, 6))
    plt.plot(fpr, tpr, color='orange', label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.title('Receiver Operating Characteristic (ROC) Curve for YOLOv8 Object Detection')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
import torch
import torchvision.transforms as T
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import requests

# Load pretrained DETR model and set to evaluation mode
model = torch.hub.load('facebookresearch/detr', 'detr_resnet50', pretrained=True)
model.eval()

# Define COCO labels for class names
COCO_CLASSES = [
    '__background__', 'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus', 'train', 'truck', 'boat',
    'traffic light', 'fire hydrant', 'stop sign', 'parking meter', 'bench', 'bird', 'cat', 'dog', 'horse',
    'sheep', 'cow', 'elephant', 'bear', 'zebra', 'giraffe', 'backpack', 'umbrella', 'handbag', 'tie', 'suitcase',
    'frisbee', 'skis', 'snowboard', 'sports ball', 'kite', 'baseball bat', 'baseball glove', 'skateboard',
    'surfboard', 'tennis racket', 'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl', 'banana',
    'apple', 'sandwich', 'orange', 'broccoli', 'carrot', 'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch',
    'potted plant', 'bed', 'dining table', 'toilet', 'tv', 'laptop', 'mouse', 'remote', 'keyboard', 'cell phone',
    'microwave', 'oven', 'toaster', 'sink', 'refrigerator', 'book', 'clock', 'vase', 'scissors', 'teddy bear',
    'hair drier', 'toothbrush'
]

# Function to preprocess the image
def preprocess_image(image_path):
    transform = T.Compose([
        T.Resize(800),
        T.ToTensor(),
        T.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
    ])
    image = Image.open(image_path)
    return transform(image).unsqueeze(0), image

# Load and preprocess the image
image_path = 'path_to_your_image.jpg'  # Replace with the path to your image
img_tensor, original_image = preprocess_image(image_path)

# Run the model on the input image
with torch.no_grad():
    outputs = model(img_tensor)

# Process the model outputs
probas = outputs['pred_logits'].softmax(-1)[0, :, :-1]  # Get class probabilities
keep = probas.max(-1).values > 0.7  # Filter by confidence threshold

# Get the bounding boxes and classes for filtered predictions
bboxes_scaled = outputs['pred_boxes'][0, keep] * torch.Tensor([original_image.width, original_image.height] * 2)
pred_classes = probas[keep].argmax(-1)

# Plot the image with detected boxes
fig, ax = plt.subplots(1, figsize=(16, 10))
ax.imshow(original_image)

# Draw bounding boxes and labels
for box, cls in zip(bboxes_scaled, pred_classes):
    x_min, y_min, x_max, y_max = box
    rect = patches.Rectangle((x_min, y_min), x_max - x_min, y_max - y_min, linewidth=2, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    ax.text(x_min, y_min, COCO_CLASSES[cls], fontsize=15, color='white', bbox=dict(facecolor='red', alpha=0.5))

plt.axis('off')
plt.figure(figsize=(12, 6))
plt.plot(fpr, tpr, color='orange', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
plt.title('Receiver Operating Characteristic (ROC) Curve for Object Detection')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
plt.grid()
plt.show()

```

```python
import cv2
import mediapipe as mp
import numpy as np
import os

# Initialize MediaPipe Objectron and drawing utilities
mp_objectron = mp.solutions.objectron
mp_drawing = mp.solutions.drawing_utils

# Define input and output paths
input_dir = 'input_images'  # Directory containing images
output_dir = 'output_images'  # Directory to save output images
os.makedirs(output_dir, exist_ok=True)

# Initialize Objectron model
with mp_objectron.Objectron(static_image_mode=True, 
                            max_num_objects=5, 
                            min_detection_confidence=0.5, 
                            model_name='Cup') as objectron:
    # Loop over each image in the directory
    for filename in os.listdir(input_dir):
        if filename.endswith('.jpg') or filename.endswith('.png'):
            image_path = os.path.join(input_dir, filename)
            image = cv2.imread(image_path)

            # Convert the image to RGB for MediaPipe processing
            image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

            # Perform object detection
            results = objectron.process(image_rgb)

            # Process and annotate detected objects
            if results.detected_objects:
                for detected_object in results.detected_objects:
                    # Draw bounding box landmarks and connections
                    mp_drawing.draw_landmarks(
                        image, 
                        detected_object.landmarks_2d, 
                        mp_objectron.BOX_CONNECTIONS, 
                        mp_drawing.DrawingSpec(color=(255, 0, 0), thickness=2),
                        mp_drawing.DrawingSpec(color=(0, 255, 0), thickness=1))
                    
                    # Draw 3D axes for each detected object
                    mp_drawing.draw_axis(
                        image, 
                        detected_object.rotation, 
                        detected_object.translation)

                    # Compute and display bounding box dimensions (approximation)
                    landmarks = detected_object.landmarks_2d.landmark
                    x_coords = [landmark.x * image.shape[1] for landmark in landmarks]
                    y_coords = [landmark.y * image.shape[0] for landmark in landmarks]
                    box_width = max(x_coords) - min(x_coords)
                    box_height = max(y_coords) - min(y_coords)

                    # Display bounding box dimensions on the image
                    cv2.putText(
                        image, 
                        f'Width: {box_width:.1f}px, Height: {box_height:.1f}px', 
                        (int(min(x_coords)), int(min(y_coords) - 10)), 
                        cv2.FONT_HERSHEY_SIMPLEX, 
                        0.5, 
                        (255, 255, 255), 
                        2)

                # Save the annotated image
                output_path = os.path.join(output_dir, filename)
                cv2.imwrite(output_path, image)
                print(f'Saved annotated image to {output_path}')
            else:
                print(f'No objects detected in {filename}')

print("Processing complete.")

    plt.axis('off')
    plt.figure(figsize=(12, 6))
    plt.plot(fpr, tpr, color='orange', label=f'ROC Curve (AUC = {roc_auc:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', linestyle='--')
    plt.title('Receiver Operating Characteristic (ROC) Curve for Object Detection')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.legend()
    plt.grid()
    plt.show()
```
